## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

conn = sqlite3.connect("m4_survey_data.sqlite")
query = "Select Age FROM master"
df = pd.read_sql_query(query, conn)
conn.close()

sns.boxenplot(data=df,y='Age')
plt.title('Boxplot of Age')
plt.ylabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

conn = sqlite3.connect("m4_survey_data.sqlite")
query = "Select Age,WorkWeekHrs FROM master"
df = pd.read_sql_query(query, conn)
conn.close()

sns.scatterplot(data=df,x='Age',y='WorkWeekHrs')
plt.title('Scatter plot of Age and Work Week Hours')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.show()
```

```python
# Correlation shenanigans
conn = sqlite3.connect("m4_survey_data.sqlite")
query = "SELECT Age, WorkWeekHrs FROM master"
df = pd.read_sql_query(query, conn)
conn.close()

# Calculate the correlation between Age and WorkWeekHrs
correlation = df['Age'].corr(df['WorkWeekHrs'])

print("Correlation between Age and WorkWeekHrs:", correlation)
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
conn = sqlite3.connect("m4_survey_data.sqlite")
query = "Select Age, WorkWeekHrs, CodeRevHrs FROM master"
df = pd.read_sql_query(query, conn)
conn.close()

plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*5, alpha=0.3)
plt.title('Bubble plot of Work Week Hours vs Code Rev Hours')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Rev Hours')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# my code goes here
conn = sqlite3.connect("m4_survey_data.sqlite")
query = "Select * FROM DatabaseDesireNextYear"
df = pd.read_sql_query(query, conn)
conn.close()

# Get the top 5 databases
top_databases = df['DatabaseDesireNextYear'].value_counts().head(5)

# Plotting a pie chart
plt.figure(figsize=(8, 6))
plt.pie(top_databases, labels=top_databases.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query the data and load it into a pandas DataFrame
query = "SELECT LanguageDesireNextYear FROM LanguageDesireNextYear"
df = pd.read_sql_query(query, conn)

# Close the connection to the database
conn.close()

# Count the occurrences of each language
language_counts = df['LanguageDesireNextYear'].value_counts()

# Find the rank of Python
python_rank = language_counts.index.get_loc('Python') + 1

print("Rank of Python among the most popular languages respondents wish to learn next year:", python_rank)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
conn = sqlite3.connect("m4_survey_data.sqlite")
query = "SELECT Age, WorkWeekHrs, CodeRevHrs FROM master WHERE Age >= 30 AND Age <= 35"
df = pd.read_sql_query(query, conn)
conn.close()

# Calculate median WorkWeekHrs and CodeRevHrs
median_workweekhrs = df['WorkWeekHrs'].median()
median_coderevhrs = df['CodeRevHrs'].median()

# Plotting a stacked bar chart
plt.figure(figsize=(10, 6))
plt.bar('30-35', median_workweekhrs, label='Median WorkWeekHrs')
plt.bar('30-35', median_coderevhrs, bottom=median_workweekhrs, label='Median CodeRevHrs')
plt.xlabel('Age Group')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.legend()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
conn = sqlite3.connect("m4_survey_data.sqlite")
query = "SELECT Age, ConvertedComp FROM master WHERE Age >= 45 AND Age <= 60"
df = pd.read_sql_query(query, conn)
conn.close()

# Group the data by age and calculate the median ConvertedComp for each age
median_convertedcomp_by_age = df.groupby('Age')['ConvertedComp'].median()

# Plotting a stacked bar chart
plt.figure(figsize=(10, 6))
plt.bar('30-35', median_workweekhrs, label='Median WorkWeekHrs')
plt.bar('30-35', median_coderevhrs, bottom=median_workweekhrs, label='Median CodeRevHrs')
plt.xlabel('Age Group')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.legend()
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
conn = sqlite3.connect("m4_survey_data.sqlite")
query = "SELECT MainBranch FROM master"
df = pd.read_sql_query(query, conn)
conn.close()

# Count the occurrences of each category
main_branch_counts = df['MainBranch'].value_counts()

# Plotting a horizontal bar chart
plt.figure(figsize=(10, 6))
main_branch_counts.plot(kind='barh', color='skyblue')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.title('Count of Main Branch')
plt.show()
```
